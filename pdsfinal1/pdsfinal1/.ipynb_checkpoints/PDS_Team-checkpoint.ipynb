## Introduction

Yelp is an American multinational corporation and it publishes crowd-sourced reviews, which allows us to get the best local content and user reviews from millions of businesses across 32 countries. 

Although Yelp has ratings for restaurants, the grading standard of users may differ and the psychological expectations of different restaurants also differ. To help users get a more credible ratings of the restaurants, we want to predict the ratings by restaurants info and comments using sentiment analysis and text classification.

In this project, we want to find out which factors has great influence on a business’ Yelp rating, and give a more general and credible rating generated by these factors. To regrade, we will construct different models trained by all restaurants data, to predict the rating for each restaurant. These ratings wouldn't hava the infleunce of different users' grading standard. In addition, we're going to split the data and train the models by category, then regrade them separately in order to avoid the bias of expectations for different kind of food. We believe our regraded ratings would give users more objective and credible evaluation of the restuarants.

### Get Restaurants

- Get the objective info of all restaurants in New York using Yelp API. 

```python
# Read the Yelp API Key from file。
def read_api_key(filepath):
    with open(filepath, 'r') as f:
        return f.read().replace('\n','')
api_key = read_api_key('api_key.txt')
```

```python
# Retrieve ALL the restaurants on Yelp for a given query.
def all_restaurants(api_key, query):
    headers = {'Authorization' : 'Bearer ' + api_key}
    parameters = {'location' : query, 'categories' : 'restaurants'}
    response = requests.get('https://api.yelp.com/v3/businesses/search', params = parameters , headers = headers)
    total = int(response.json()['total']) if 'total' in response.json() else 0
    data = response.json()['businesses'] if 'businesses' in response.json() else []
    offset = 20
    while offset < total:
        time.sleep(.200)
        parameters = {'location' : query, 'categories' : 'restaurants', 'offset' : offset}
        response = requests.get('https://api.yelp.com/v3/businesses/search', params = parameters , headers = headers)
        data += (response.json()['businesses'] if 'businesses' in response.json() else [])
        offset += 20
    return data
```

```python
# Get all resestaurants in New York.
ny_restaurants = all_restaurants(api_key, 'New York')
```

### Retrieve Restaurants Info

- Retrieve URLs from previous data then get the raw HTML of each restaurant.

```python
# Parse Yelp API results to extract restaurant URLs.
def parse_api_response(data):
    dataDict = json.loads(data)
    urls = []
    for business in dataDict['businesses']:
        urls.append(business['url'])
    return urls  
```

```python
# Return the raw HTML at the specified URL.
def retrieve_html(url):
    response = requests.get(url)
    return (response.status_code, response.text)
```

- Parse the HTML page of a restaurant. Retrieve the first 50 reviews for a restaurant on Yelp.

```python
# Parse the reviews on a single page of a restaurant.
def parse_page(html):
    bs = BeautifulSoup(html, 'html.parser')
    reviews = []
    for lis in bs.find_all('ul', {'class' : 'ylist ylist-bordered reviews'}):
        for element in lis.find_all('li', recursive = False):
            review = {}
            review_id = None
            if 'js-war-widget' in element.div['class']:
                continue
            review_id = element.div['data-review-id']
            user_id = element.div['data-signup-object']
            rating = float(element.find('img', {'class' : 'offscreen'})['alt'][:3])
            date = element.find('span', {'class' : 'rating-qualifier'}).string
            text = element.find('div', {'class' : 'review-content'}).find('p')

            if review_id and user_id and rating is not None and date and text:
                review['review_id'] = review_id
                review['user_id'] = user_id.split(':')[1]
                review['rating'] = rating
                review['date'] = date.strip()
                review['text'] = text.getText()
                reviews.append(review)
    nextPageTag = bs.find('a', {'class' : 'u-decoration-none next pagination-links_anchor'})
    nextPageURL = nextPageTag['href'] if nextPageTag else None
    return reviews, nextPageURL
```

```python
# Retrieve the first 50 reviews for a single restaurant on Yelp.
def extract_reviews(url):
    reviews = []
    code, html = retrieve_html(url)
    tmpReviews, nextPageURL = parse_page(html)
    index = 0
    while nextPageURL != None:
        reviews += tmpReviews
        code, html = retrieve_html(nextPageURL)
        tmpReviews, nextPageURL = parse_page(html)
        index += len(tmpReviews)
        if (index > 50):
            break
    return reviews + tmpReviews
```

- Get all the info we need as features and generate the records.

```python
# Collect all info needed from raw restaurants data.
def collect_info(restaurants):
    res = []
    for r in restaurants:
        info = {}
        info['name'] = r['name']
        info['transactions'] = r['transactions']
        info['categories'] = r['categories']
        info['price'] = r['price']
        info['review_count'] = r['review_count']
        info['city'] = r['location']['city']
        info['state'] = r['location']['state']
        info['rating'] = r['rating']
        info['reviews'] = extract_reviews(r['url'])
        res.append(info)
    return res
```

```python
# Collect info of the first 1000 restaurants.
records = collect_info(ny_restaurants[:1000])
```

### Load Data to Database

- Create a sqlite database and design tables to store the data we got.

```python
# Create tables in sqlite database.
def create_tables(cursor):
    cursor.execute('''CREATE TABLE restaurants
             (name text, transactions text, categories text, city text, state text, review_count int, price text, rating real)''')
    cursor.execute('''CREATE TABLE reviews
             (restaurant_name text, categories text, review_rating real, comment text)''')
```

```python
# Export data to sqlite database.
def export_to_db(cursor, records):
    for record in records:
        name = record['name']
        cats = []
        for cat in record['categories']:
            cats.append(cat['alias']) 
        rt = (name, '\t'.join(record['transactions']), '\t'.join(cats), record['city'], record['state'], record['review_count'], record['price'], record['rating'])
        cursor.execute('''INSERT INTO restaurants VALUES (?,?,?,?,?,?,?,?)''', rt)
        for review in record['reviews']:
            pt = (name, review['rating'], review['text'])
            cursor.execute('''INSERT INTO reviews VALUES (?,?,?)''', pt)
```

- Load the data of 1000 New York restaurants to sqlite database.

```python
# Create a sqlite database.
def create_db(db_name, records):
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    create_tables(cursor)
    export_to_db(cursor, records)
    conn.commit()
    conn.close()
```

```python
# Load the New York restaurants data to database.
create_db('ny_restaurants_1000.db', records)
```

## Sentiment Analysis Regression Models

### Generate Features

Besides those features (category, transaction, price, review count) we get directly from Yelp, we also create some other features from reviews. In this project, we use two methods, that is generating positive, negative and compund score given a review; and tfidf matrix. For each review text, we assign one or more scores to it, and use the average scores as the features of each restaurant. For the first method, we usually use compound and positive score; for the second one, we will discuss it in detail later.

For regression models, we use the average review rating scores as the labels instead of the rating for every restaurant. Because the overall star rating is an approximation to the average review rating, it is not accurate if we use this in regression models.

Intuitively, reviews is an important factor. To prove this, two categories of features are used in the regression models, one only including transaction, review_count and price, the other including compund and positive scores generated from reviews besides those features in the first category. Baseline models use the first category of features and some other models use the second category.

### Sentiment Analysis Features

- Generate sentiment features using NLTK api. This api would give 3 scores (positive, negative and compound) to represent the sentiment for a sentence.

```python
# Sentiment analysis.Get the positive, negative and compound scores given a review.
def get_sentiment_scores(text):
    sentences = tokenize.sent_tokenize(text)
    sid = SentimentIntensityAnalyzer()
    compounds = []
    positives = []
    negatives = []
    for sentence in sentences:
        ss = sid.polarity_scores(sentence)
        compounds.append(ss['compound'])
        positives.append(ss['pos'])
        negatives.append(ss['neg'])
    return np.mean(compounds), np.mean(positives), np.mean(negatives)
```

- Combine the objective features and sentiment features and create the dataframe including all features and labels needed for future processing. The 'real rating' is calculated by the average ratings of reviews for a restaurant. Since we only select 50 reviews for a restaurant but not all reviews, the 'real rating' would be more accurate than the rating of the whole restaurant.

```python
# Add the sentiment features.
def add_sentiment_features(df):
    compounds = []
    real_ratings = []
    positives = []
    negatives = []
    for i in range(df.shape[0]):
        name = df.iloc[i]['name']
        query = 'SELECT comment, review_rating FROM reviews WHERE restaurant_name = "%s";' % name
        reviews_df = pd.read_sql_query(query , con)
        compound = 0
        positive = 0
        negative = 0
        real_rating = 0
        if reviews_df.shape[0] == 0:
            continue
        for j in range(reviews_df.shape[0]):
            scores = get_sentiment_scores(reviews_df.iloc[j]['comment'])
            compound += scores[0]
            positive += scores[1]
            negative += scores[2] 
            real_rating += reviews_df.iloc[j]['review_rating']
        compounds.append(compound / reviews_df.shape[0])
        positives.append(positive / reviews_df.shape[0])
        negatives.append(negative / reviews_df.shape[0])
        real_ratings.append(real_rating / reviews_df.shape[0])
    df = df.assign(real_rating=pd.Series(np.array(real_ratings)))
    df = df.assign(compound=pd.Series(np.array(compounds)))
    df = df.assign(positive=pd.Series(np.array(positives)))
    df = df.assign(negative=pd.Series(np.array(negatives)))
    return df
```

```python
# Add the sentiment analysis features to the dataframe
df = add_sentiment_features(df)
```

### Regression by Different Models

```python
# SVR with 'linear' kernel
def SVR_linear(X_train, y_train, X_test, y_test):
    rlf = svm.SVR(kernel='linear', C=1.0)
    rlf.fit(X_train, y_train)
    predictions = rlf.predict(X_test)
    return predictions, rlf.score(X_test, y_test)
```

```python
# SVR with 'sigmoid' kernel
def SVR_sigmoid(X_train, y_train, X_test, y_test):
    rsf = svm.SVR(kernel='sigmoid', C=1.0)
    rsf.fit(X_train, y_train)
    predictions = rsf.predict(X_test)
    return predictions, rsf.score(X_test, y_test)
```

```python
# SVR with rbf kernel
def SVR_rbf(X_train, y_train, X_test, y_test):
    rbf = svm.SVR(kernel='rbf', C=1.0)
    rbf.fit(X_train, y_train)
    predictions = rbf.predict(X_test)
    return predictions, rbf.score(X_test, y_test)
```

```python
# Decision Tree Regression
from sklearn import tree
def DTR(X_train, y_train, X_test, y_test):
    clf = tree.DecisionTreeRegressor()
    clf = clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    return predictions, clf.score(X_test, y_test)
```

```python
# Ridge Regression
def RidgeReg(X_train, y_train, X_test, y_test):
    rdg = Ridge()
    rdg = rdg.fit(X_train, y_train)
    predictions = rdg.predict(X_test)
    return predictions, rdg.score(X_test, y_test)
```

```python
# Kernel Ridge Regression
def KernelRidgeReg(X_train, y_train, X_test, y_test):
    kr = KernelRidge()
    kr = kr.fit(X_train, y_train)
    predictions = kr.predict(X_test)
    return predictions, kr.score(X_test, y_test)
```

## Sentiment Analysis Classification Models

Since reviews have great influence on a business' rating, we will construct two kinds of classifiers. One kind of models use only features created from reviews and the other kind uses not only features created from reviews, but also numerical features like transaction, review count and price.

Continuous labels generated above are not suitable foe classification models any more, here we use the overall rating for each business from Yelp. The labels are from one to five with an increment of 0.5.

Classification models are evaluated by accuracy rate, because labels are not binary and we only focus on whether the predicted label is consistent with the true one.

```python
# Get features and labels for classification.
def classification_generate_features_labels(df):
    X = np.zeros((df.shape[0], 6))
    Y = np.zeros((df.shape[0]))
    for i in range(df.shape[0]):
        restaurant = df.iloc[i]
        X[i][0] = len(restaurant['transactions'].split('\t')) 
        X[i][1] = restaurant['review_count']
        X[i][2] = len(restaurant['price'])
        X[i][3] = restaurant['compound']
        X[i][4] = restaurant['positive']
        X[i][5] = restaurant['negative']
        Y[i] = restaurant['rating'] * 2
    return X, Y
```

```python
# Get the features and labels for Classification. And split them to train and test sets.
cX, cY = classification_generate_features_labels(df[:-1])
cX_train, cX_test, cy_train, cy_test = train_test_split(cX, cY, test_size=0.2)
```

- Different classification models, including decision tree, random forest, SVC with linear, sigmoid and rbf kernel.

```python
# SVC with linear kernel
def SVC_linear(X_train, y_train, X_test, y_test):
    clf = svm.SVC(kernel='linear', C=1)
    clf.fit(X_train, y_train)
    return clf.score(X_test, y_test)
```

```python
# SVC with sigmoid kernel
def SVC_sigmoid(X_train, y_train, X_test, y_test):
    clf = svm.SVC(kernel='sigmoid', C=1)
    clf.fit(X_train, y_train)
    return clf.score(X_test, y_test)
```

```python
# SVC with rbf kernel
def SVC_rbf(X_train, y_train, X_test, y_test):
    clf = svm.SVC(kernel='rbf', C=1)
    clf.fit(X_train, y_train)
    return clf.score(X_test, y_test)
```

```python
# Decision Tree
def DecisionTree(X_train, y_train, X_test, y_test):
    dt = DecisionTreeClassifier(criterion='gini', max_depth=25)
    dt.fit(X_train, y_train)
    return dt.score(X_test, y_test)
```

```python
# Random Forest
def RandomForest(X_train, y_train, X_test, y_test):
    clf = RandomForestClassifier(n_estimators=200)
    clf.fit(X_train, y_train)
    return clf.score(X_test, y_test)
```

For the last prediction, we get a list of blocks of indexes. Each block represents a restaurant and it contains indexes of reviews of this restaurant.

```python
# get the index of restaurants, when we predict a certain restaurant, we should know which reviews belong to which restaurant
def create_df_by_restaurants(restaurant_group, restaurants):
    index = 0
    res = []
    restaurant_index_lst = []
    for restaurant in restaurants:
        small_group = restaurant_group.get_group(restaurant)
        restaurant_index_lst.append(list(range(index, index+len(small_group))))
        res.append(small_group)
        index += len(small_group)
    
    return  pd.concat(res), restaurant_index_lst
    
```

### Generate features

The next step is to use tfidf method to create features from our reviews. The first thing we do is to process reviews by converting to lower case, replacing "'s" "'" with "", eliminating punctuations, tokenizing messages and lemmatizing words.

```python
# preprocess data, create tfidf features
def process(message, lemmatizer=nltk.stem.wordnet.WordNetLemmatizer()):
    message = message.lower()                          #1. transfer all tokens to lower case
    message = message.replace("'s", "")                #2. replace all "'s" with ""
    message = message.replace("'", "")                 #3. replace all "'" with ""
    for p in string.punctuation:                  
        message = message.replace(p, " ")              #4. eliminate all punctuations
    res = []
    word_list = nltk.word_tokenize(message)            #5. tokenize message to reduce dimensions and redundancy
    for word in word_list:
        try:
            word = lemmatizer.lemmatize(word)          #6. lemmatize all words
            res.append(word)
        except e:
            pass 
    return res
```

```python
def process_all(df, lemmatizer=nltk.stem.wordnet.WordNetLemmatizer()):
    df['comment'] = df['comment'].apply(lambda x:process(x))
    return df

df = process_all(df, lemmatizer=nltk.stem.wordnet.WordNetLemmatizer())
```

The we get rare words of which the number is 1

```python
def get_rare_words(df):
    size = df.shape[0]
    rare_word_lst = []
    for line in df.comment:
        rare_word_lst.extend(line)   
    
    rare_word_dict = Counter(rare_word_lst)  
    res = [key for key, val in rare_word_dict.items() if val == 1]
    return sorted(res)

rare_words = get_rare_words(df)  
```

And we create our tfidf feature matrix by eliminating stop words, rare words. And we get different lists of words to use.

```python
def create_features(df, rare_words):
    
    stopwords = nltk.corpus.stopwords.words('english')  # stopwords to omit
    nomeanwords = []
    
    meaningless_words = stopwords + rare_words + nomeanwords         # allwords including stopwords and rarewords to omit
    meaningless_words = set(meaningless_words)
    size = df.shape[0]
    text_lst = []                                       # a list consisting of all words
    clean_text_lst = []                                 # a list consisting of all words that omit stopwords and rarewords
    one_clean_text_lst = []                             # a list consisting of all words that omit stopwords and rarewords that belong to ham
    two_clean_text_lst = []                            # a list consisting of all words that omit stopwords and rarewords that belong to spam
    three_clean_text_lst = []
    four_clean_text_lst = []
    five_clean_text_lst = []
    
    
    
    for i in df.index.values:
        text_lst.append(" ".join(df.comment[i]))
        if df['review_rating'][i] == 1:
            for word in df.comment[i]:
                if word not in meaningless_words:
                    clean_text_lst.append(word)
                    one_clean_text_lst.append(word)
        elif df['review_rating'][i] == 2:
            for word in df.comment[i]:
                if word not in meaningless_words:
                    clean_text_lst.append(word)
                    two_clean_text_lst.append(word)     
        elif df['review_rating'][i] == 3:
            for word in df.comment[i]:
                if word not in meaningless_words:
                    clean_text_lst.append(word)
                    three_clean_text_lst.append(word)  
        elif df['review_rating'][i] == 4:
            for word in df.comment[i]:
                if word not in meaningless_words:
                    clean_text_lst.append(word)
                    four_clean_text_lst.append(word)  
        else:
            for word in df.comment[i]:
                if word not in meaningless_words:
                    clean_text_lst.append(word)
                    five_clean_text_lst.append(word)  
            
                
    tfid = sklearn.feature_extraction.text.TfidfVectorizer(stop_words = meaningless_words)
    csr = tfid.fit_transform(text_lst)
    return tfid, csr, text_lst, clean_text_lst, one_clean_text_lst,two_clean_text_lst,three_clean_text_lst,four_clean_text_lst,five_clean_text_lst


(tfid, csr, text_lst, clean_text_lst, one_clean_text_lst,two_clean_text_lst,three_clean_text_lst,four_clean_text_lst,five_clean_text_lst
) = create_features(df, rare_words)
```

### Applying Text Classification Model to Regrade Restaurants

Then we can use our best model random forest(trained by all kinds of categories) to regrade our restaurants. For example, we get all reviews of a restaurant,and use our model to get the rating of each of review. Then we get the average score of these reviews to represent the rating of this restaurant and compare to is original score to see the difference. As a result, the difference is about 28%.

```python
# predict rating for all reviews
preds = random_forest_clf.predict(csr)
```

```python
# average review rating for certain restaurant
predict_res = []
for lst in restaurant_index_lst:
    res = []
    for i in lst:
        
        res.append(preds[i])
    predict_res.append(np.mean(res))
```

```python
# classify those ratings 
pred_class = []
for pred in predict_res:
    if pred >= 4.75: pred_class.append(5)
    elif pred >=4.25: pred_class.append(4.5)
    elif pred >= 3.75: pred_class.append(4)
    elif pred >= 3.25: pred_class.append(3.5)
    elif pred >= 2.75: pred_class.append(3)
    elif pred >= 2.25: pred_class.append(2.5)
    elif pred >= 1.75: pred_class.append(2)
    elif pred >= 1.25: pred_class.append(1.5)
    else: pred_class.append(1)
```

```python
# compare those ratings with original ratings
error = 0
for i in range(len(pred_class)):
    if pred_class[i] != rating_lst[i]:
        error += 1
difference = (error / len(pred_class))
print(difference)

#0.2811
```
